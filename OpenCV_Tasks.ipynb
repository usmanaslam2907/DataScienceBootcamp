# Brightness the image

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

path = '/content/Image.jpg'
image = cv2.imread(path)
image_float=image.astype(np.float32)
brighter_image = np.clip(image_float + 50, 0, 255).astype(np.uint8)

cv2_imshow(image)
print('\n')
cv2_imshow(brighter_image)
```

# Rotation the image

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
path = '/content/Image.jpg'
image = cv2.imread(path)
height, width = image.shape[:2]
center = (width // 2, height // 2)
for angle in range(0, 361, 10):
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale=1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    cv2_imshow(rotated_image)

```

# Car object Detection tasks
